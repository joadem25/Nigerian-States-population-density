Collate the following informations for every 36 states in Nigeria.
    State Name
    State Capital
    State Population
    State Landmass
Create a Pandas Series object of state population and landmass using state name as the key of both.
From the two Series objects, create a DataFrame object and add the state capital as new collumn to the table.
Calculate the population density of each state and use it to generate a new collumn on the DataFrame object.

```python
# Collate the following informations for every 36 states in Nigeria.
#         State Name
#         State Capital
#         State Population
#         State Landmass
stateName = ['Abia', 'Adamawa', 'Akwa Ibom', 'Anambra', 'Bauchi', 'Bayelsa', 'Benue', 'Borno', 'Cross River', 'Delta', 
             'Ebonyi', 'Edo', 'Ekiti', 'Enugu', 'Gombe', 'Imo', 'Jigawa', 'Kaduna', 'Kano', 'Katsina', 'Kebbi', 'Kogi', 
             'Kwara', 'Lagos', 'Nasarawa', 'Niger', 'Ogun', 'Ondo', 'Osun', 'Oyo', 'Plateau', 'Rivers', 'Sokoto', 'Taraba',
             'Yobe', 'Zamfara']
stateCapital =['Umuahia', 'Yola', 'Uyo', 'Awka', 'Bauchi', 'Yenagoa', 'Makurdi', 'Maiduguri', 'Calabar', 'Asaba', 
               'Abakaliki', 'Benin', 'Ado Ekiti', 'Enugu', 'Gombe', 'Owerri', 'Dutse', 'Kaduna', 'Kano', 'Katsina', 
               'Birnin Kebbi', 'Lokoja', 'Ilorin', 'Ikeja', 'Lafia', 'Minna', 'Abeokuta', 'Akure', 'Oshogbo', 'Ibadan', 
               'Jos', 'Port Harcourt', 'Sokoto', 'Jalingo', 'Damaturu', 'Gusau']
statePopulation = [3841943, 4536948, 4780581, 5599910, 7540663, 2394725, 5787706, 5751590, 4175020, 5307543, 3007155, 
                   4461137, 3350401, 4396098, 3623462, 5167722, 6779080, 8324285, 14253549, 9300382, 5001610, 4153734, 
                   3259613, 12772884, 2632239, 6220617, 5945275, 4969707, 4237396, 7512855, 4400974, 7034973, 5863187, 
                   3331885, 3398177, 5317793]
stateLandmass = [6320, 36917, 7081, 4844, 45893, 10773, 34059, 70898, 20156, 17698, 5670, 17802, 6353, 7161, 18768, 5530, 
                 23154, 46053, 20131, 24192, 36800, 29833, 36825, 3345, 27117, 76363, 16762, 15500, 9251, 28454, 30913, 
                 11077, 25973, 54473, 45502, 39762]
print('State Names\n', stateName, '\n')
print('State Capitals\n', stateCapital, '\n')
print('State Population\n', statePopulation, '\n')
print('State Landmass\n', stateLandmass)
```

```python
# Create a Pandas Series object of state population and landmass using state name as the key of both
# name = np.array(ngaData['State Name'])
population = pd.Series(statePopulation, index=stateName)
landmass = pd.Series(stateLandmass, index=stateName)
print("Nigerian State's population\n",population, '\n')
print("Nigerian State's landmass\n", landmass)
```

```python
# From the two Series objects, create a DataFrame object and add the state capital as new collumn to the table.
nig_data = pd.DataFrame({'Population':population, 'Landmass':landmass})
nig_data['Capital'] = stateCapital
nig_data
```

```python
# Calculate the population density of each state and use it to generate a new collumn on the DataFrame object.
nig_data['Population density'] = nig_data['Population'] / nig_data['Landmass']
nig_data
```
